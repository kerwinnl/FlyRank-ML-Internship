# 1. My Lane as an ML Task (Type)

**Task Type:** Ranking

This project is a **ranking** task because the goal is to prioritize content pages that should be refreshed first. Rather than predicting a simple yes/no outcome, the model produces a ranked list of content items according to their likelihood of declining in performance.

**Decision:** Which content pages should editors update first?

**Who uses the output?** Content editors and SEO specialists.

**Action:** Refresh or optimize the highest-priority pages before their performance declines further.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

# 2. Target or Proxy

The target for this assignment is **is_declining_label**, which indicates whether a content item is declining.

According to the dataset documentation, **trend_direction** and **trend_pct** are derived from this label and therefore must not be used as input features because they would introduce data leakage.

# 3. Success Metric

The primary evaluation metric is **Precision@K**.

Editors usually have limited time and can only update a small number of pages. Precision@K measures how many of the highest-ranked pages are actually declining, making it an appropriate metric for content prioritization.

# 4. Why ML Beats a Fixed Rule

A simple rule such as "CTR below 1%" or "traffic dropped by 20%" cannot capture the complex relationships among multiple performance signals.

Content performance depends on many factors, including clicks, impressions, search position, engagement, and content type. These factors interact in ways that are difficult to express using a single rule.

Machine learning can learn these complex patterns from historical data and produce a more useful ranking for editors. The model supports decision-making rather than replacing human judgment.
